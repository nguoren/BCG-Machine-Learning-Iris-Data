## Data Examination

```python
import pandas as pd

# import datasets from statsmodels.api
# get dataset from sklearn
from sklearn import datasets

iris = datasets.load_iris()
print(iris)
```

```python
iris.data
```

```python
iris.target
```

```python
iris.target_names
```

```python
# create dataframe for the iris dataset
iris_df = pd.DataFrame(iris['data'], columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])
iris_df
```

```python
# create dataframe for the target value 
iris_target_df = pd.DataFrame(iris['target'], columns = ['Target'])
iris_target_df
```

```python
# merge iris dataframe and target value dataframe
iris_df['Target'] = iris_target_df
iris_df
```

```python
# map species names onto iris dataframe
iris_target_names_dict = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
iris_df['Species'] = iris_df['Target'].map(iris_target_names_dict)
iris_df
```

```python
# check the data type of iris dataframe
iris_df.dtypes
```

```python
# get a count of the different species in the dataframe
iris_df.groupby('Species').count()
```

#### Remarks:
- There are a total of 150 samples with 50 samples in each species of _**Setosa**_, _**Versicolor**_ and _**Virginica**_

## Data Preparation

```python
# scale the numeric values of the features
from sklearn.preprocessing import scale
num_cols = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
iris_scaled_df = scale(iris_df[num_cols])
iris_scaled_df = pd.DataFrame(iris_scaled_df, columns = num_cols)
iris_scaled_df
```

```python
iris_scaled_df.describe().round(3)
```

```python
# assign numeric value to represent the species
iris_scaled_df['Species'] = iris_df['Target']
iris_scaled_df
```

## K-Nearest Neighbors (KNN) Model

```python
# train_test_split the dataset
from sklearn.model_selection import train_test_split
import numpy as np
np.random.seed(3456)
iris_split = train_test_split(np.asmatrix(iris_scaled_df), test_size = 75)
iris_train_features = iris_split[0][:, :4]
iris_train_labels = np.ravel(iris_split[0][:, 4])
iris_test_features = iris_split[1][:, :4]
iris_test_labels = np.ravel(iris_split[1][:, 4])
print(iris_train_features.shape)
print(iris_train_labels.shape)
print(iris_test_features.shape)
print(iris_test_labels.shape)
```

```python
# define and train the KNN model
from sklearn.neighbors import KNeighborsClassifier
KNN_mod = KNeighborsClassifier(n_neighbors = 3)
KNN_mod.fit(iris_train_features, iris_train_labels)
```

```python
# test dataframe
iris_test = pd.DataFrame(iris_test_features, columns = num_cols)
iris_test
```

```python
# predicted results of the test dataframe by the KNN model
iris_test['prediction'] = KNN_mod.predict(iris_test_features)
iris_test
```

```python
iris_test_labels
```

```python
# add actual answers onto the dataframe for comparison
iris_test['actual'] = iris_test_labels
iris_test
```

```python
# add column to show whether the prediction is correct (1) / incorrect (0)
iris_test['correctness'] = np.where(iris_test['prediction'] == iris_test['actual'], 1, 0)
iris_test
```

```python
# calculate the accuracy of the model
accuracy = float(sum(iris_test['correctness']) / (iris_test['correctness']).count())
print(f'The accuracy of the KNN Model is {accuracy*100}%')
```
